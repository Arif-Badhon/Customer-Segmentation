```python
for col in df.columns:
    temp = df[col].isnull().sum()
    if temp > 0:
        print(f'Column {col} contains {temp} null values.')
```

```python
df = df.dropna()
print("Total missing values are:", len(df))
```

```python
df.nunique()
```

```python
parts = df["Dt_Customer"].str.split("-", n=3, expand=True)
df["day"] = parts[0].astype('int')
df["month"] = parts[1].astype('int')
df["year"] = parts[2].astype('int')
```

```python
df.drop(['Z_CostContact', 'Z_Revenue', 'Dt_Customer'],
        axis=1,
        inplace=True)
```

```python
floats, objects = [], []
for col in df.columns:
    if df[col].dtype == object:
        objects.append(col)
    elif df[col].dtype == float:
        floats.append(col)

print(objects)
print(floats)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 7))

for i, col in enumerate(objects):
    # Generate a palette with as many colors as unique values in the column
    unique_categories = df[col].unique()
    palette = sns.color_palette("husl", len(unique_categories))
    
    # Plot using the custom palette
    sns.countplot(x=df[col], ax=axes[i], palette=palette)
    axes[i].set_title(f'Count plot for {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
#LabelEncoding
for col in df.columns:
    if df[col].dtype == object:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
```

```python
#Heatmap
plt.figure(figsize=(15, 15))
sns.heatmap(df.corr() > 0.8, annot=True, cbar=False)
plt.show()
```
